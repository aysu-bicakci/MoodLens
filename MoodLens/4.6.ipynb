```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture(0)
```

```python
while True:
    ret, frame = cap.read()
    if not ret:
        break

    try:
        # Dataset içindeki yüzlerle karşılaştırma
        result = DeepFace.find(img_path=frame, db_path="dataset", enforce_detection=False)

        if len(result[0]) > 0:
            identity_path = result[0].iloc[0]["identity"]
            name = identity_path.split("\\")[-2]  # klasör adı kişiyi verir
            cv2.putText(frame, f"{name}", (30, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        else:
            cv2.putText(frame, "Kisi bulunamadi", (30, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    except Exception as e:
        cv2.putText(frame, "Yuz algilanamadi", (30, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
        print("Hata:", e)

    cv2.imshow("Kamera", frame)

    # 'q' ya basarsan çıkar
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()
```

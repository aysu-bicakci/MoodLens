```python
import cv2
from tensorflow.keras.models import load_model
import numpy as np
import time
```

```python
# Eğittiğiniz modeli yükleyin
model_path = 'model_dropout.h5'
model = load_model(model_path)
```

```python
# Webcam'den görüntü almak için bir capture objesi oluşturun
cap = cv2.VideoCapture(0)
```

```python
# FPS (kare hızı) sınırlayıcıları
fps_limit = 5  # Her saniyede en fazla 5 kare işleyeceğiz
interval = 1.0 / fps_limit  # Güncelleme aralığı hesaplanıyor
 
last_time = time.time()  # Başlangıç zamanını alın
```

```python
while True:
    ret, frame = cap.read()  # Webcam'den bir frame alın
 
    # Her saniyede belirli bir kare hızı ile tahminleri güncelleyin
    if time.time() - last_time >= interval:
        last_time = time.time()  # Zamanı güncelleyin
 
        # Görüntüyü model için uygun formata getirin
        resized_frame = cv2.resize(frame, (48, 48))
        img_array = np.expand_dims(resized_frame, axis=0) / 255.0
 
        # Eğitilen modeli kullanarak tahmin yapın
        prediction = model.predict(img_array)
 
        # Tahmin sonucunu işleyin
        prob_happy = prediction[0][0] * 100
        prob_sad = (1 - prediction[0][0]) * 100
 
        if prediction[0][0] > 0.5:
            label = "Happy: {:.2f}% | Sad: {:.2f}%".format(prob_happy, prob_sad)
        else:
            label = "Happy: {:.2f}% | Sad: {:.2f}%".format(prob_happy, prob_sad)
 
        # Etiketi ve çerçeveyi görselleştirin
        cv2.putText(frame, label, (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        cv2.imshow('Emotion Detection', frame)
 
    if cv2.waitKey(1) & 0xFF == ord('q'):  # Çıkış için 'q' tuşuna basın
        break
 
# Döngüden çıkınca işlemleri serbest bırakın ve pencereyi kapatın
cap.release()
cv2.destroyAllWindows()
```

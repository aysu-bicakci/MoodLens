```python
import cv2
from deepface import DeepFace

# Video dosyasını aç
video_path = "aysu_video.mp4"  # Buraya videonun bulunduğu yolu yaz
cap = cv2.VideoCapture(video_path)

while True:
    ret, frame = cap.read()  # Videodan bir kare al
    if not ret:
        break  # Eğer kare alınamazsa, döngüden çık

    # Yüzleri tanımak için DeepFace kullan
    try:
        result = DeepFace.analyze(frame, actions=['face_detection'], enforce_detection=False)
        print(result)  # Tanıdık yüzler ve diğer özellikler

        # Yüzleri etiketle
        for face in result[0]["region"]:
            x, y, w, h = face["x"], face["y"], face["w"], face["h"]
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    except Exception as e:
        print(f"Yüz tanıma hatası: {e}")

    # Videodaki kareyi göster
    cv2.imshow("Video Yüz Tanıma", frame)

    # 'q' tuşuna basıldığında çık
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# İşlem bittikten sonra kaynakları serbest bırak
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import joblib
from deepface import DeepFace
from keras.models import load_model
import time

# Yüz tanıma için Cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# KNN Modelini Yükle
knn_model = joblib.load("dataset/face_knn_model.pkl")

# Duygu tahmini için model
emotion_model = load_model("model_dropout.h5")

# Video dosyasını oku (kamera yerine dosya yolu veriyoruz)
cap = cv2.VideoCapture("tilki.mp4")  # ← buraya video dosyanın ismini yaz

fps_limit = 5
interval = 1.0 / fps_limit
last_time = time.time()

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    current_time = time.time()
    if (current_time - last_time) < interval:
        continue
    last_time = current_time

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

    for (x, y, w, h) in faces:
        face_img = frame[y:y+h, x:x+w]
        
        # 1. Yüz Tanıma
        try:
            result = DeepFace.represent(img_path=face_img, model_name="Facenet", enforce_detection=False)
            embedding = result[0]['embedding']
            embedding = np.expand_dims(embedding, axis=0)

            prediction = knn_model.predict(embedding)
            name = str(prediction)
            cv2.putText(frame, f"Name: {name}", (x, y + h + 25), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 200, 0), 2)
        except Exception as e:
            print(f"Yüz Tanıma Hatası: {e}")

        # 2. Duygu Analizi
        try:
            resized_face = cv2.resize(face_img, (48, 48))
            resized_face = np.expand_dims(resized_face, axis=0)
            resized_face = resized_face / 255.0
            prediction = emotion_model.predict(resized_face)
            emotion_label = "Sad" if prediction[0][0] > 0.5 else "Happy"
            emotion_text = f"{emotion_label}: {max(prediction[0][0], 1 - prediction[0][0]) * 100:.1f}%"
            cv2.putText(frame, emotion_text, (x, y + h + 50), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 200, 0), 2)
        except Exception as e:
            print(f"Duygu Analizi Hatası: {e}")

    # Göster
    cv2.imshow("Video", frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import joblib
from deepface import DeepFace
from keras.models import load_model
import time

# Haar Cascade ile yüz tanıma
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# KNN Modeli yükleniyor
knn_model = joblib.load("face_knn_model.pkl")

# Duygu tahmin modeli yükleniyor
emotion_model = load_model("model_dropout.h5")

# Video dosyasını aç (kamera yerine dosya yolu girilebilir)
cap = cv2.VideoCapture("tilki_video.mp4")  # ← buraya video dosyanı koy

# FPS düşürmek için zaman kontrolü
fps_limit = 5
interval = 1.0 / fps_limit
last_time = time.time()

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    current_time = time.time()
    if (current_time - last_time) < interval:
        continue
    last_time = current_time

    # Gri yerine renkli bırakıyoruz
    faces = face_cascade.detectMultiScale(frame, scaleFactor=1.1, minNeighbors=5)

    for (x, y, w, h) in faces:
        face_img = frame[y:y+h, x:x+w]

        # ---------- 1. YÜZ TANIMA ----------
        try:
            rgb_face = cv2.cvtColor(face_img, cv2.COLOR_BGR2RGB)
            result = DeepFace.represent(img_path=rgb_face, model_name="Facenet", enforce_detection=False)
            embedding = np.expand_dims(result[0]['embedding'], axis=0)

            prediction = knn_model.predict(embedding)
            name = prediction[0]

            cv2.rectangle(frame, (x, y), (x + w, y + h), (100, 255, 100), 2)
            cv2.putText(frame, f"Name: {name}", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 200, 0), 2)

        except Exception as e:
            print(f"Yüz Tanıma Hatası: {e}")

        # ---------- 2. DUYGU ANALİZİ ----------
        try:
            resized_face = cv2.resize(face_img, (48, 48))
            resized_face = resized_face / 255.0
            resized_face = np.expand_dims(resized_face, axis=0)

            prediction = emotion_model.predict(resized_face)
            emotion_label = "Sad" if prediction[0][0] > 0.5 else "Happy"
            emotion_score = max(prediction[0][0], 1 - prediction[0][0]) * 100
            emotion_text = f"{emotion_label}: {emotion_score:.1f}%"

            cv2.putText(frame, emotion_text, (x, y + h + 25), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 255), 2)

        except Exception as e:
            print(f"Duygu Analizi Hatası: {e}")

    # Sonucu ekranda göster
    cv2.imshow("Video", frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

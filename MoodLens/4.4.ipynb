```python
import os
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
```

```python
# Veri yollarını tanımlayalım
train_dir = "C:\\Users\\Yusuf\\source\\jupyter\\dl\\images\\train"
validation_dir = "C:\\Users\\Yusuf\\source\\jupyter\\dl\\images\\validation"
```

```python
# Veri artırma ve ölçekleme
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

validation_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(48, 48),
    batch_size=32,
    class_mode='binary'
)

validation_generator = validation_datagen.flow_from_directory(
    validation_dir,
    target_size=(48, 48),
    batch_size=32,
    class_mode='binary'
)
```

```python
# CNN modeli oluşturalım
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(48, 48, 3)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten(),
    Dropout(0.5),  # Dropout ekleyelim
    Dense(512, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# Modeli eğitelim
history_combined = model.fit(
    train_generator,
    steps_per_epoch=200,
    epochs=50,
    validation_data=validation_generator,
    validation_steps=50
)
```

```python
# Sonuçları çizdirelim
train_acc_combined = history_combined.history['accuracy']
val_acc_combined = history_combined.history['val_accuracy']
epochs_combined = range(1, len(train_acc_combined) + 1)

plt.plot(epochs_combined, train_acc_combined, 'bo', label='Training accuracy with Dropout and Data Augmentation')
plt.plot(epochs_combined, val_acc_combined, 'b', label='Validation accuracy with Dropout and Data Augmentation')
plt.title('Training and validation accuracy with Dropout and Data Augmentation')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# Modeli kaydedelim
model.save('model_dropout_and_augmentation.h5')
```
